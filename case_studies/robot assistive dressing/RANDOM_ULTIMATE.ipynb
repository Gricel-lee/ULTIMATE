```python
import random

def gen_dat_file(fileName, N, Nmin, Nmax, mean):
    # Set seed for reproducibility
    random.seed(42)

    # Generate N random numbers between ...
    numbers = [random.uniform(Nmin, Nmax) for _ in range(N)]

    # Adjust numbers to achieve the target mean of 20
    current_mean = sum(numbers) / len(numbers)
    difference = mean - current_mean

    # Adjust the values to bring the mean closer to 20
    adjusted_numbers = [min(Nmax, max(Nmin, num + difference)) for num in numbers]
    # Rounded to 3 decimal places
    adjusted_numbers = [round(num, 2) for num in adjusted_numbers]

    # Ensure the mean is exactly 20 by rounding the sum if necessary
    final_mean = sum(adjusted_numbers) / len(adjusted_numbers)

    # Output the N numbers as a comma separated string
    print(adjusted_numbers)
    print(f"Final Mean: {final_mean}")

    # Save to file
    with open(fileName, "w") as f:
        f.write(",".join(map(str, adjusted_numbers)))
```

```python
import random

def generate_binary_array(N, mean, fileName, seed=42):
    # Set seed for reproducibility
    random.seed(seed)
    
    # Calculate the number of 1's required
    num_ones = int(N * mean)
    
    # Create an array with num_ones 1's and the rest 0's
    binary_array = [1] * num_ones + [0] * (N - num_ones)
    
    # Shuffle the array to randomize the positions of 1's and 0's
    random.shuffle(binary_array)
    
    # Save to file
    with open(fileName, "w") as f:
        f.write(",".join(map(str, binary_array)))
        
    mean_value = sum(binary_array) / len(binary_array)
    # print(binary_array)
    print(f"Mean: {mean_value}")

    return binary_array

```

```python
fileName = "___p1_ok_correct.dat"
N = 500
Nmin = 0
Nmax = 1
mean = 0.7
gen_dat_file(fileName, N, Nmin, Nmax, mean)
```
